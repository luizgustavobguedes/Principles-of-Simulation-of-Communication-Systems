```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000 # Quantidade de amostras geradas
U = np.random.uniform(0,1,N) # Geração de variáveis aleatórias com distribuição uniforme
count = 0
lambda1 = 1/28

x_gen = -np.log(U)/(lambda1) # Valores de x gerados pelo método da inversa
x = np.arange(0,np.max(x_gen),0.01)
x_t = lambda1*np.exp(-lambda1*x) # PDF teórica da distribuição exponencial

# Probabilidades calculadas de acordo com a condição e seguindo o conceito de frequência relativa
for k in x_gen:
  if k <= 4:
    count = count + 1

p = count/N

print("Probabilidade de se ter quatro chamadas ou menos:")
print(p)

plt.plot(x,x_t,color = 'black')
plt.hist(x_gen, bins = 100, density = 'True', color = 'blue', label = 'Distribuição exponencial')
plt.xlabel("Amostras")
plt.ylabel("Histograma")
plt.legend()
plt.show()
```
