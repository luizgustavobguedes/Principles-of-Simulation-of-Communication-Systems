```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda1 = 3 # Número médio de requisições
N = 100000 # Numero de amostras
count = 0
av = np.array([])
x = np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)

for k in av:
  if k >= 2:
    count = count + 1

p_a = count/N

print("Vetor de amostras gerado:")
print(av)
print("Probabilidade de se ter mais de duas falhas:")
print(p_a)

plt.hist(av, bins = 50, density = True, color = "red", label = "Histograma normalizado da distribuição de Poisson")
plt.xlabel("Amostras")
plt.ylabel("Histograma")
plt.legend()
plt.show()
```
