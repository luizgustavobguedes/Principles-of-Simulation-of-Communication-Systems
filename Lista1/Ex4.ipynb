```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda1 = 3 # Número médio de requisições
N = 100000 # Numero de amostras
count = 0
av = np.array([])
x = np.random.uniform(0,1,N) # Geração de um vetor de variáveis aleatórias com distribuição Uniforme
for ix in x: # Para um dado valor desse vetor uniforme
    i = 0 # Inicializa-se a variável i com o valor 0 (i representa o valor correspondente ao valor da variável aleatória seguindo distribuição de Poisson)
    pr = np.exp(-lambda1) # Probabilidade inicial
    F=pr # A probabilidade acumulada é a probabilidade inicial
    while ix>=F: # Enquanto o valor do vetor uniforme for maior ou igual à probabilidade acumulada
        pr=lambda1/(i+1)*pr # Calcula-se a probabilidade de um elemento seguinte
        F = F + pr # Acumula-a
        i = i + 1; # Incrementa a variável i a cada necessidade de se acumular, ou seja, até a acumulada tornar-se maior que o valor de probabilidade do vetor uniforme
    a1=i # Qaudno a probabilidade do vetor uniforme for menor que a probabilidade acumulada, salva-se o valor de i e armazena-o em a1
    av=np.append(av,a1) # Anexa-se a1 no vetor que conterá variáveis aleatórias seguindo distribuição de Poisson

# Cálculo da probabilidade de acordo com a condição e seguindo o conceito de frequência relativa
for k in av:
  if k >= 2:
    count = count + 1

p_a = count/N

print("Vetor de amostras gerado:")
print(av)
print("Probabilidade de se ter mais de duas falhas:")
print(p_a)

plt.hist(av, bins = 50, density = True, color = "red", label = "Histograma normalizado da distribuição de Poisson")
plt.xlabel("Amostras")
plt.ylabel("Histograma")
plt.legend()
plt.show()
```
