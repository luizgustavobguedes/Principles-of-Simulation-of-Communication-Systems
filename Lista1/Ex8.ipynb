```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100000
xfunction = np.array([])

for i in range(n): # Experimento a ser realizado n vezes
  x1=np.random.uniform(-1,1) # Gera-se uma variável aleatória com distribuição uniforme nos limites desejados
  x2=np.random.uniform(0,1) # Gera-se uma variável aleatória com distribuição uniforme entre 0 e 1

# Lembrando que x1^2 foi calculado analiticamente pelo método da aceitação-rejeição
  while (x2 >=pow(x1,2)): # Enquanto o valor de x2 for maior ou igual a x1^2
    x1=np.random.uniform(-1,1) # Gera-se outra variável aleatória com distribuição uniforme nos limites desejados
    x2=np.random.uniform(0,1) # Gera-se outra variável aleatória com distribuição uniforme entre 0 e 1
  xfunction = np.append(xfunction, x1) # Quando x2 for menor que o x1^2, pega-se aceita-se esse valor e armazena no vetor almejado

print("O vetor de variáveis aleatórias formado é este: ")
print(xfunction)
X=np.arange(-1, 1, 0.1)
fx=1.5*pow(X,2)
plt.plot(X,fx)
plt.hist(xfunction,bins=100,density=True)
plt.xlabel("Amostras")
plt.ylabel("Histograma")
plt.show()
```
