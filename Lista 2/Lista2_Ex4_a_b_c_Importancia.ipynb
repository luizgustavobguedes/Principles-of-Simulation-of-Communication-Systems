```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000000

u = np.random.uniform(0,1,N)

x1 = pow(u,1/3)

integral1 = np.sum(pow((1-pow(x1,2)),3/2)/(3*pow(x1,2)))/N
print("Pelo método da importância, o resultado da integral da letra a) é: ")
print(integral1)

x = np.linspace(0, 1, N)
curve_y = pow((1-pow(x,2)),(3/2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(0.5, 0.8, f'Integral = {integral1:.4f}', fontsize=12, color='red')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
N = 1000000

u = np.random.uniform(0,1,N)

x2 = np.log(u*(np.exp(2)-np.exp(-2))+np.exp(-2))

integral2 = np.sum(np.exp(x2+pow(x2,2))/(np.exp(x2)/(np.exp(2)-np.exp(-2))))/N

print("Pelo método da importância, o resultado da integral da letra b) é: ")
print(integral2)

x = np.linspace(-3, 3, N)
curve_y = np.exp(x+pow(x,2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(0.5, 0.8, f'Integral = {integral2:.4f}', fontsize=12, color='red')
plt.xlim(-3, 3)
plt.ylim(0, 15)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
N = 1000000

u = np.random.uniform(0,1,N)

y = np.sqrt(u)
x3 = (1-y)/y

Z = x3/pow((1+pow(x3,2)),2)/(2*y*pow(y,2))

integral3 = np.sum(Z)/N

print("Pelo método da importância, o resultado da integral da letra c) é: ")
print(integral3)

x = np.linspace(0, 10, N)
curve_y = x*pow((1+pow(x,2)),-2)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(0.5, 0.8, f'Integral = {integral3:.4f}', fontsize=12, color='red')
plt.xlim(0, 10)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
