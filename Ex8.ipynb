```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100000
xfunction = np.array([])

for i in range(n):
  x1=np.random.uniform(-1,1)
  x2=np.random.uniform(0,1)

  while (x2 >=pow(x1,2)):
    x1=np.random.uniform(-1,1)
    x2=np.random.uniform(0,1)
  xfunction = np.append(xfunction, x1)

print(xfunction)
X=np.arange(-1, 1, 0.1)
fx=1.5*pow(X,2)
plt.plot(X,fx)
plt.hist(xfunction,bins=100,density=True)
plt.show()
```
